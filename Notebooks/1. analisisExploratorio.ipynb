### üìÇ Carga de Datos  

En esta secci√≥n se realizar√° la carga de los datos que se utilizar√°n para el an√°lisis de informaci√≥n donde se exploraran los datos del **Super Bowl**

Las fuentes de los datos son **3 archivos en formato CSV**, los cuales contienen la informaci√≥n relacionada con audiencia de television, los resultados de los juegos y los espectaculos de medio tiempo

### Archivos CVS utilizados
1. `halftime_musicians.csv`
2. `super_bowls.csv`
3. `tv.csv`


### üéØ Objetivo
- Unificar los archivos en un **DataFrame**
- Realizar la **limpieza de los datos** y validar la informaci√≥n
- Preparar los datos para el **an√°lisis exploratorio inicial**

```python
import pandas  as pd
from IPython.display import display, Markdown
```

```python
#Carga de los datos
df_halftime_musicians = pd.read_csv('../Data/Raw/halftime_musicians.csv')
df_super_bowls = pd.read_csv('../Data/Raw/super_bowls.csv')
df_tv = pd.read_csv('../Data/Raw/tv.csv')

```

```python
#Validar la carga de los datos
display(Markdown("**DATOS ESPECTACULOS DE MEDIO TIEMPO**".center(100,'-')))
display(df_halftime_musicians.head())
```

```python
#Validar la carga de los datos
display(Markdown("**DATOS RESULTADOS SUPER BOWLS**".center(100,'-')))
display(df_super_bowls.head())


```

```python
display(Markdown("**DATOS NULOS DEL SUPER BOWLS**".center(100,'-')))
df_super_bowls.isnull().sum()

```

```python
display(Markdown("**ELIMINAR LAS COLUMNAS DEL SUPER BOWLS**".center(100,'-')))
df_super_bowls = df_super_bowls.drop(columns=['qb_winner_2','qb_loser_2' ])
df_super_bowls.isnull().sum()
```

```python
#Validar la carga de los datos
display(Markdown("**DATOS AUDIENCIA TELEVISI√ìN**".center(100,'-')))
display(df_tv.head())
```

```python
df_tv.isnull().sum()
```

```python
#Filtrar los datos nulo
df_tv[df_tv.isnull().any(axis=1)]
```

```python
#Elmimnar las columnas
df_tv = df_tv.drop(columns=['total_us_viewers','rating_18_49','share_18_49' ])
df_tv.isnull().sum()
```

##### üóÇÔ∏è **TIPOS DE DATOS Y NOMBRE DE LAS COLUMNAS DEL DATAFRAME**
Para explorar la estructura de un DataFrame en **Pandas**, es importante identificar tanto los nombres de las columnas como el tipo de dato que contiene cada una de ellas. Esto permite verificar si la informaci√≥n est√° correctamente representada

```python
display(Markdown("**INFORMACI√ìN ESPECTACULOS DE MEDIO TIEMPO**".center(100,'-')))
df_halftime_musicians.info()

display(Markdown("**INFORMACI√ìN RESULTADOS DEL SUPER BOWLS**".center(100,'-')))
df_super_bowls.info()

display(Markdown("**INFORMACI√ìN AUDIENCIA TV**".center(100,'-')))
df_tv.info()
```

### üß© Uni√≥n de Tablas y Plan de Limpieza de Datos

Para iniciar el an√°lisis de ventas junto con los datos de los clientes, se realiza una **uni√≥n de las tablas `halftime_musicians` , `super_bowls` , `tv` usando `outer join`**. Esto permite trabajar con **toda la informaci√≥n disponible** antes de hacer la limpieza.  


```python
df = pd.merge(df_super_bowls, df_tv, on='super_bowl', how='outer')
df = pd.merge(df, df_halftime_musicians, on='super_bowl', how='outer')
display(Markdown("**INFORMACI√ìN DATAFRAME UNIFICADO**".center(100,'-')))
df.info()
```

```python
display(df.head())
```

```python
# Validar los datos nulos 
display(Markdown("**VALORES NULOS POR COLUMNA**".center(100,'-')))
null_counts = df.isnull().sum()
print(null_counts)
```

```python
#Identificar las filas duplicadas
duplicates = df[df.duplicated()]
print("N√∫mero de filas duplicadas:", len(duplicates))
```

```python
#Filtrar los datos nulo
df[df.isnull().any(axis=1)]
```

```python
#Exportar el dataframe limpio
df.to_csv('../Data/Processesd/super_bowls_clean.csv', index=False)
```
